```python
import time
from Scweet.scweet import scrape
import datetime
```

```python
start_time = '2021-11-15'
end_time = '2022-02-15'
while 1:
    data = scrape(words=["Meta", "meta", "Facebook", "facebook"], since=start_time, until=end_time, from_account=None,
                interval=1,
                headless=False, display_type="Latest", save_images=False, save_dir='outputs',
                resume=False, filter_replies=True, proximity=False)
    s_date = datetime.datetime.strptime(start_time, '%Y-%m-%d').date() + datetime.timedelta(days=1)
    start_time = s_date.strftime("%Y-%m-%d")
    if start_time == end_time:
        print("Finished")
        break
    time.sleep(1)
```
